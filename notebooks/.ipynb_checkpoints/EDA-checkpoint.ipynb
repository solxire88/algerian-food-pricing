# Library Imports

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# Loading Data

```python
food = pd.read_csv("../data/processed/food.csv", index_col=0)
food.head()
```

```python
indicators = pd.read_csv("../data/processed/indicators.csv", index_col=0)
indicators.head()
```

```python
food.index = pd.to_datetime(food.index)
indicators.index = pd.to_datetime(indicators.index)
```

# Exploratory Data Analysis

```python
fig, ax = plt.subplots(2,2, figsize=(12,8))
ax = ax.flatten()

ax[0].set_title("Count by Province")
ax[0].set_xlabel(" ")
sns.countplot(data=food, x='province', ax=ax[0])
ax[1].set_title("Count by Town")
ax[1].set_xlabel(" ")
sns.countplot(data=food, x='town', ax=ax[1])
ax[2].set_title("Count by Market")
ax[2].set_xlabel(" ")
sns.countplot(data=food, x='market', ax=ax[2])
ax[3].set_title("Count by Category")
ax[3].set_xlabel(" ")
ax[3].tick_params(axis='x', rotation=45)
sns.countplot(data=food, x='category', ax=ax[3])
```

```python
food.groupby(['date', 'category'])['price'].mean().unstack('category').sort_index().plot(figsize=(16,10), subplots=True, layout= (3,3),
                                                                                      title='Average Price of Food in each category',
                                                                                      ylabel='Price (DZD)')
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))
fig.suptitle("Impact of Agricultural Land Reduction on Cereal and Vegetable Prices Over Time", fontsize=16, fontweight='bold')
food.groupby(['date', 'category'])['price'].mean().unstack('category')[["cereals and tubers", "vegetables and fruits"]].sort_index().plot(ax=ax[0], title="Average Price (in local currency or USD)", xlabel="", sharex=True)
indicators["Agricultural land (% of land area)"].plot(ax=ax[1], xlabel="", rot=30, sharex=True, title='Agricultural Land (% of Total Land Area)')
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle(
    "Impact of Agricultural Land Reduction on Cereal and Vegetable Prices Over Time",
    fontsize=16, fontweight='bold'
)

price_ts = (
    food
    .groupby(['date', 'category'])['price']
    .mean()
    .unstack('category')
    [["cereals and tubers", "vegetables and fruits"]]
    .sort_index()
)
price_ts.plot(
    ax=ax[0],
    color=['#1f77b4', '#ff7f0e'],
    marker='o',
    markersize=4,
    title="Average Price (DZD)",
    xlabel=""
)
ax[0].legend(title='Food Category', loc='upper left', framealpha=0.9)

indicators["Agricultural land (% of land area)"].plot(
    ax=ax[1],
    marker='s',
    markersize=4,
    title="Agricultural Land (% of Total Land Area)",
    xlabel="",
    rot=30
)

for a in ax:
    a.grid(which='major', linestyle='--', alpha=0.5)
    a.spines['top'].set_visible(False)
    a.spines['right'].set_visible(False)


ax[0].axvspan(pd.Timestamp('2016-01-01'), pd.Timestamp('2022-01-01'),
              color='gray', alpha=0.2)
ax[1].axvspan(pd.Timestamp('2016-01-01'), pd.Timestamp('2022-01-01'),
              color='gray', alpha=0.2)

fig.text(
    0.5, 0.01,
    'Data sources: World Food Programme & United States Group On Earth Observations',
    ha='center', fontsize=9, style='italic'
)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
